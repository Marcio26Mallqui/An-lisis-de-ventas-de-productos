```python
import pandas as pd
import glob
```

```python
lista_de_archivos = glob.glob(r'C:\Users\HP\Desktop\Trabajos para prender\Proyectos personales\Proyectos de datos de Marcio Mallqui\Ventas -python - sql - PowerBi\datos/*csv')
# Conectarnos a la carpeta que contiene los archivos.csv  
```

### Unimos los datos de todos los archivos csv

```python
# Unir todos los datos que se encuentran en los archivos
df_total = pd.DataFrame()
for archivos in lista_de_archivos:
    df1 = pd.read_csv(archivos, sep = ',["]*', engine = 'python')
    df_total = pd.concat([df_total, df1])
```

```python
df_total = df_total.reset_index()
df_total
```

```python
df_total = df_total.rename(columns={'level_0': 'product', 'level_1': 'quantity_ordered', 'Product' : 'price_each', 
                                    'Quantity Ordered': 'order_date', 'Price Each': 'street_number_name', 	'Order Date' : 'city', 
                                    'Purchase Address': 'state_zip_code', 'Category': 'category'})
df_total
```

```python
# separamos la columna state_zip_code
df_total[['space', 'state', 'zip_code']] = df_total['state_zip_code'].str.split(' ', expand= True)
df_total
```

### Uniendo columnas

Hay que tener en cuenta que la forma de escribir una dirección en Estados Unidos, donde los estadounidenses escriben la ciudad, estado y código postal (incluye todo esto en una línea con una coma entre ciudad y estado, pero no el código postal).



Esta información los puede encontrar en la siguiente página web:

https://www.campussims.com/how-to-write-a-us-address/
Esto puede ser debido a que hay ciudades que tienen el mismo nombre ubicadas en diferentes estados del país de Estados Unidos.U.

#### **En este caso solo vamos unir la ciudad y el estado, no tomaremos en cuenta el código postal.**

```python
# unión de columnas
df_total['city_state'] = df_total['city'].str.cat(df_total['state'], sep = '/ ')
df_total
```

```python
# Eliminamos las columnas que no necesitamos
df_total = df_total.drop(columns= [	'city',	'state_zip_code', 'space', 'state', 'zip_code'])
df_total
```

### Eliminando caracteres no deseados

```python
# limpiar las comillas
df_total.iloc[:, [0,5]] = df_total.iloc[:, [0,5]].apply(lambda x: x.str.strip('"'))
# El método apply() de Pandas es un método de propósito general que se puede utilizar 
# para aplicar una función a cada elemento de una serie o cada fila o columna de un marco de datos (DataFrame).

# str.strip(), remover caracteres que no deseamos.
df_total
```

```python
# Los valores  únicos de cada columna
for columna in df_total.columns:
    print(df_total[columna].unique())
```

```python
# Eliminar el espacio del inicio en la columna city_state

df_total.iloc[:, [6]] = df_total.iloc[:, [6]].apply(lambda x: x.str.strip(' '))
# El método apply() de Pandas es un método de propósito general que se puede utilizar 
# para aplicar una función a cada elemento de una serie o cada fila o columna de un marco de datos (DataFrame).

# str.strip(), remover caracteres que no deseamos.
df_total
```

```python
# Los valores  únicos de cada columna
for columna in df_total.columns:
    print(df_total[columna].unique())
# desaparecio el espacio    
```

```python
# eliminamos los datos duplicados
df_total = df_total.drop_duplicates()
```
